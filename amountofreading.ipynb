```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 생성 또는 로드 (예시 데이터)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct'],
    'Fiction': [5, 7, 6, 8, 10, 9, 7, 6, 5, 4],
    'Non-Fiction': [3, 4, 6, 5, 4, 7, 8, 6, 5, 4],
    'Sci-Fi': [2, 3, 4, 3, 2, 1, 2, 4, 3, 2],
    'Self-Help': [4, 5, 3, 6, 7, 5, 4, 5, 6, 7]
}

df = pd.DataFrame(data)

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 층별로 데이터를 쌓아 그래프 그리기
plt.stackplot(df['Month'], df['Fiction'], df['Non-Fiction'], df['Sci-Fi'], df['Self-Help'], labels=['Fiction', 'Non-Fiction', 'Sci-Fi', 'Self-Help'], alpha=0.7)

plt.title('월별 독서 그래프', fontsize=16)
plt.xlabel('월', fontsize=14)
plt.ylabel('월 간 읽은 책 수', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 범례 추가
plt.legend(loc='upper left', fontsize=12)

# 그래프 저장 또는 표시
plt.tight_layout()
plt.savefig('reading_graph.png', dpi=300)
plt.show()
```
